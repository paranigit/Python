```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# In order to see the visualization inline use the below magic commend; Only for Jupyter notebook
%matplotlib inline
```

```python
x = np.arange(0,10)
y = x**2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, "*")
```

```python
plt.plot(x,y,"red")
```

```python
plt.plot(x,y,"r--")
plt.xlim(0,5)
plt.ylim(0,30)
plt.title("sample title")
plt.xlabel("nums")
plt.ylabel("squares")
```
